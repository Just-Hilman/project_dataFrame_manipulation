## Question 4

```python
#Menammbahkan kolom baru dengan nama Month 
data0.insert(1, 'Month', data0['Paid Date'])
```

```python
#Mengubah format kolom Month menjadi %Y%m, karena saat ini formatnya masih %Y%m%d dan merubah tipe datanya menjadi object
data0['Month'] = pd.to_datetime(data0['Month'], format="%Y%m%d").dt.strftime("%Y%m")
```

```python
#Mengubah tipe data kolom Month di data3 kedalam object sehingga dapat dibandingkan dengan kolom Month yang ada didalam data0 ketika dilakukan merge, karena saat ini tipe datanya masih int
data3['Month'] = pd.to_datetime(data3['Month'], format="%Y%m").dt.strftime("%Y%m")
```

```python
#Melakukan merge antara data0 dan data3 dengan kolom yang dibandingkan adalah Month dan Meta Category sehingga muncul sebuah kolom baru yaitu Transaction Fee Rate yang nantinya akan dilakukan perhitungan untuk mendapatkan Transaction Fee Amount(Transaction AMount * Transaction Fee Rate)
datamerge = pd.merge(data0, data3, left_on=['Month', 'Meta Category'], right_on=['Month', 'Meta Category'], how='left')
datamerge
```

```python
#Perhitungan untuk mendapatkan Transaction Fee Amount
trx_fee_amount = datamerge['Transaction Fee Rate'] * datamerge['Transaction Amount']
trx_fee_amount
```

```python
#Melakukan penambahan kolom baru yaitu Net Revenue sesuai yang diminta pada soal, dimana Net Revenue adalah hasil perhitungan antara Transaction Fee Amount - Sales Discount - Other Discount
datamerge.insert(16, 'Net Revenue', trx_fee_amount - datamerge['Sales Discount'] - datamerge['Other Discount'])
datamerge
```

## Question 5

```python
#Menambahkan kolom baru bernama %Net Revenue, dimana %Net Revenue adalah hasil perhitungan antara Net Revenue / GMV
datamerge.insert(17, '%Net Revenue', datamerge['Net Revenue'] / datamerge['GMV'])
datamerge
```

```python
#Melakukan groupby untuk mengetahui rata-rata Monthly GMV
monthly_gmv = datamerge.groupby(['Month']).agg({'GMV':'mean'})
monthly_gmv.columns = ['Monthly GMV']
monthly_gmv
```

```python
#Melakukan groupby untuk mengetahui rata-rata untuk monthly %Net revenue
monthlyNetRevenue_percentage = datamerge.groupby(['Month']).agg({'%Net Revenue':'mean'})
monthlyNetRevenue_percentage.columns = ['Monthly %Net Revenue']
monthlyNetRevenue_percentage
```

```python
#Melakukan merge Monthly GMV dengan Monthly %Net Revenue untuk dapat divisualisasikan menggunakan chart
plotmerge = pd.merge(monthly_gmv, monthlyNetRevenue_percentage, left_index=True, right_index=True)
plotmerge
```

```python
from matplotlib import pyplot as plt
```

```python
#Melakukan visualisasi data yang didapat dengan bar chart untuk Monthly GMV dan line chart untuk Monthly %Net Revenue
fig, ax = plt.subplots()
plotmerge.plot(
    use_index=True, 
    kind='bar',
    y='Monthly GMV', 
    ax=ax, 
    color='orange'
)

plotmerge.plot(
    use_index=True, 
    y='Monthly %Net Revenue', 
    ax=ax, 
    secondary_y=True, 
    color='blue'
)

plt.show()
```

## Question 8

```python
#Melakukan merge pada data0 dengan data1 berapa lama pembeli dari mulai melakukan pembayaran hingga melakukan konfirmasi
numberOfDays = pd.merge(data0, data1, left_on=['Order Number'], right_on=['Order Number'], how='left')
#Merubah dan menyamakan tipe data Paid Date dengan Purchase Confirmation Date menjadi datetime, karena saat ini tipe datanya masih int dan akan menghasilkan data yang tidak valid ketika dilakukan perhitungan selisih hari
numberOfDays['Paid Date'] = pd.to_datetime(numberOfDays['Paid Date'],format="%Y%m%d")
numberOfDays['Purchase Confirmation Date'] = pd.to_datetime(numberOfDays['Purchase Confirmation Date'],format="%Y%m%d")
```

```python
#Menambahkan kolom baru dengan nama Confirmation Days Taken untuk memudahkan mengetahui selisih hari yang diperlukan dari mulai tanggal pembayaran hingga tanggal konfirmasi
numberOfDays.insert(17, 'Confirmation Days Taken', numberOfDays['Purchase Confirmation Date'] - numberOfDays['Paid Date'])
```

```python
numberOfDays.head()
```

```python
#Melakukan pengelompokan minimum, rata-rata, dan maksimum hari yang dibutuhkan dari mulai pembayaran hingga konfirmasi berdasarkan Meta Category pada bulan November
dayTaken = numberOfDays[(numberOfDays['Month'] == '201711')]
dayTaken.groupby(['Month', 'Meta Category']).agg({'Confirmation Days Taken':['min', 'max', 'mean']})
```

## Question 9

```python
#Melakukan merge pada data0 dengan data1 untuk mengetahui Paid Date dan Purchase Confirmation Date
merge1 = pd.merge(data0, data1, left_on=['Order Number'], right_on=['Order Number'], how='left')
```

```python
#Melakukan merge hasil dari merge yang pertama untuk mengetahui Cancel Stakeholder dan Cancel Reason
trxMonthlyRecap = pd.merge(merge1, data2, left_on=['Order Number'], right_on=['Order Number'], how='left')
trxMonthlyRecap.head(20)
```

```python
#Melakukan perhitungan Paid Transaction, %Confirm Transaction dan %Canceled Transaction
recapCount = trxMonthlyRecap.groupby(['Month']).agg({'Order Number':'count', 'Purchase Confirmation Date':'count'})
recapCount.columns = ['#Paid Transaction', 'Confirmed Transaction']
recapCount['Cancelled Trx'] = recapCount['#Paid Transaction'] - recapCount['Confirmed Transaction']
recapCount['%Paid Trx'] = recapCount['#Paid Transaction'] / recapCount['#Paid Transaction'] * 100
recapCount['%Confirmed Trx'] = recapCount['Confirmed Transaction'] / recapCount['#Paid Transaction'] * 100
recapCount['%Cancelled Trx'] = recapCount['Cancelled Trx'] / recapCount['#Paid Transaction'] * 100
recapCount
```
